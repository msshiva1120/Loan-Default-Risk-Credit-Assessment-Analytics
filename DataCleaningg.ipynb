### **1. Converting the  Loan Date to proper date format.**

```python
df.columns=df.columns.str.replace('Loan Date (DD/MM/YYYY)','Loan Date')
```

```python
df.info()
```

```python
df.columns
```

```python
df["Loan Date"].isnull().sum()
```

```python
df["Loan Date"]=df["Loan Date"].str.split('\r').str[0]
```

```python
df.head()
```

```python
df["Loan Date"]=pd.to_datetime(df["Loan Date"]).dt.date
```
